```python
import numpy as np

apples_sales = np.array([15, 22, 18, 14, 20, 25, 19])
bananas_sales = np.array([10, 12, 8, 15, 9, 11, 7])
cherries_sales = np.array([5, 7, 6, 4, 8, 10, 6])

math_scores = np.array([92, 88, 95, 90, 87])
science_scores = np.array([75, 80, 70, 85, 78])
history_scores = np.array([88, 92, 85, 90, 87])
english_scores = np.array([80, 82, 78, 85, 88])
cs_scores = np.array([95, 90, 92, 88, 94])

monthly_sales = np.array([150, 120, 135, 145, 160, 175, 170, 180, 165, 155, 140, 130])

temperatures = np.array([25, 28, 24, 26, 27, 30, 29])

fuel_efficiency = np.array([25, 28, 23, 26, 30])

print("Fruits sales data:")
print("Apples:", apples_sales)
print("Bananas:", bananas_sales)
print("Cherries:", cherries_sales)

print("\nStudent's test scores:")
print("Mathematics:", math_scores)
print("Science:", science_scores)
print("History:", history_scores)
print("English:", english_scores)
print("Computer Science:", cs_scores)

print("\nCompany's monthly sales data:")
print(monthly_sales)

print("\nTemperatures for the week:")
print(temperatures)

print("\nCar's daily fuel efficiency:")
print(fuel_efficiency)
```

# Task 6:

```python
import numpy as np
import matplotlib.pyplot as plt

x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
x_range_2pi = np.linspace(-2*np.pi, 2*np.pi, 100)
x_range_sqrt23 = np.linspace(-np.sqrt(23), np.sqrt(23), 100)
x_range_6_5_8_5 = np.linspace(-6.5, 8.5, 100)
x_range_17_3 = np.linspace(-17/3, 17/3, 100)

X, Y = np.meshgrid(x_range, y_range)
X_2pi, Y_2pi = np.meshgrid(x_range_2pi, x_range_2pi)
X_sqrt23, Y_sqrt23 = np.meshgrid(x_range_sqrt23, x_range_sqrt23)
X_6_5_8_5, Y_6_5_8_5 = np.meshgrid(x_range_6_5_8_5, x_range_6_5_8_5)
X_17_3, Y_17_3 = np.meshgrid(x_range_17_3, x_range_17_3)

Z1 = X**2 + Y**2
Z2 = X**2 - Y**2
Z3 = X**3 + Y**3
Z4 = X**2 * Y
Z5 = X**2 + Y**3

fig = plt.figure(figsize=(18, 5))

ax1 = fig.add_subplot(151, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='viridis')
ax1.set_title('Z = X^2 + Y^2')

ax2 = fig.add_subplot(152, projection='3d')
ax2.plot_surface(X_2pi, Y_2pi, Z2, cmap='plasma')
ax2.set_title('Z = X^2 - Y^2')

ax3 = fig.add_subplot(153, projection='3d')
ax3.plot_surface(X_sqrt23, Y_sqrt23, Z3, cmap='magma')
ax3.set_title('Z = X^3 + Y^3')

ax4 = fig.add_subplot(154, projection='3d')
ax4.plot_surface(X_6_5_8_5, Y_6_5_8_5, Z4, cmap='inferno')
ax4.set_title('Z = X^2 * Y')

ax5 = fig.add_subplot(155, projection='3d')
ax5.plot_surface(X_17_3, Y_17_3, Z5, cmap='plasma')
ax5.set_title('Z = X^2 + Y^3')

plt.tight_layout()
plt.show()
```
